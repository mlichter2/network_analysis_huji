# Network properties 

```python
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

upload the texts csv file from the data folder

```python
df = pd.read_csv('../data/texts.csv')
df.head()
```

```python
G = nx.Graph()
G.add_edges_from(zip(df['from'], df['to']))
```

```python
nx.draw(G, with_labels = True)
```

### Connectivity in undirected network

```python
g = {0: { 2, 3},
         1: {0},
         2: {3},
         3: { 1},
         4: {6},
         5: {3},
   6: {5},
   7: {2},
   8: {4, 9},
   9: {4},
   }
G2 = nx.DiGraph(g)
nx.draw(G2, with_labels = True)
```

##### strongly connected network

is G2 strongly connected? use `nx.is_strongly_connected(G2)` to find out

##### weakly connected network

is G2 weakly connected? use `nx.is_weakly_connected(G2)` to find out

##### Strongly connected component

get all the strongly connected components in G2 using `list(nx.strongly_connected_components(G2))`

##### Weakly connected component

get all the weakly connected components in G2 using `list(nx.weakly_connected_components(G2))`
