# Network properties 

```python
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

upload the texts csv file from the data folder

```python
df = pd.read_csv('../data/texts.csv')
df.head()
```

```python
G = nx.Graph()
G.add_edges_from(zip(df['from'], df['to']))
```

```python
nx.draw(G, with_labels = True)
```

## Clustering Coefficient

### Local clustering coefficient

find the local clustering coefficient of at least two central and peripheral nodes from both networks using `nx.clustering(G, node)`

##### texts network

```python

print('periphery', nx.clustering(G, 'Angel'), nx.clustering(G, 'Zachary'))
print('center', nx.clustering(G, 'Grace'), nx.clustering(G, 'Alexandra'))
```

##### karate club network

```python
print('periphery', nx.clustering(Gk, 14), nx.clustering(Gk, 29))
print('center', nx.clustering(Gk, 0), nx.clustering(Gk, 31))
```

### Connectivity in undirected network

```python
g = {0: { 2, 3},
         1: {0},
         2: {3},
         3: { 1},
         4: {6},
         5: {3},
   6: {5},
   7: {2},
   8: {4, 9},
   9: {4},
   }
G2 = nx.DiGraph(g)
nx.draw(G2, with_labels = True)
```

##### weakly connected network

## Network robustness

##### number of nodes needed to break the network

use `nx.node_connectivity(G)` to get the minimum number of nodes needed to break the texts  and karate club networks - which one is more robust?

```python
print('texts', nx.node_connectivity(G), 'karate',nx.node_connectivity(Gk))
```

##### Get the nodes needed to cut the network

use `nx.minimum_node_cut(G)` to get the the nodes needed to break the texts  and karate club networks - which are these nodes?

```python
print('texts', nx.minimum_node_cut(G), 'karate',nx.minimum_node_cut(Gk))
```

##### number of edges needed to break the network

use `nx.edge_connectivity(G)` to get the minimum number of edges needed to break the texts  and karate club networks - which one is more robust?

```python
print('texts', nx.edge_connectivity(G), 'karate',nx.edge_connectivity(Gk))
```

##### Get the edges needed to cut the network

use `nx.minimum_edge_cut(G)` to get the the edges needed to break the texts  and karate club networks - which are these nodes?

```python
print('edges', nx.minimum_edge_cut(G), 'edges',nx.minimum_edge_cut(Gk))
```
