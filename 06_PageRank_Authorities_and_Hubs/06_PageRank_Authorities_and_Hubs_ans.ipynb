write a function that return a dictionary containing the sorted lists (descending) of names of the `n` highest ranking nodes for the following measures:
<ol>
   
<li> degree centrality - key name 'dc'</li> 
<li> closeness centrality - key name 'cc'</li> 
<li> betweenness centrality - key name 'bc'</li> 
<li> pagerank - key name 'pr</li> 
<li> hits authorities - key name 'ha'</li> 
<li> hits hubs - key name - 'hh'</li> 
    </ol>
    
  <ul>
   
<li> the function should allow to change PageRank alpha</li> 
<li> the function should allow to define betweenness centrality endpoints and normalization</li> 
<li> if the network is undirected - return `None` instead of the hits hubs and authorities</li> 
</ul>


compare centrality measures for the texts network - representing it once as a directed and then as an undirected graph

```python
def top_n_centrality_measures(G,n = 5,  pagerank_alpha = 0.85, bc_endpoints = True, bc_normalized = True):
    measures = {}
    dc = nx.degree_centrality(G)
    dcc = sorted(tuple(zip(dc.keys(), dc.values())),key=lambda tup: tup[1], reverse=True)[:n]
    dcc = [i[0] for i in dcc]
    
    
    cc = nx.closeness_centrality(G)
    ccc = sorted(tuple(zip(cc.keys(), cc.values())),key=lambda tup: tup[1], reverse=True)[:n]
    ccc = [i[0] for i in ccc]
    
    
    bc = nx.betweenness_centrality(G, endpoints = bc_endpoints, normalized = bc_normalized)
    bcc = sorted(tuple(zip(bc.keys(), bc.values())),key=lambda tup: tup[1], reverse=True)[:n]
    bcc = [i[0] for i in bcc]
    
    pr = nx.pagerank(G, alpha = pagerank_alpha)
    prc = sorted(tuple(zip(pr.keys(), pr.values())),key=lambda tup: tup[1], reverse=True)[:n]
    prc = [i[0] for i in prc]
    
    if G.is_directed():
    
        hits = nx.hits(G)
        hits_h = sorted(tuple(zip(hits[0].keys(), hits[0].values())),key=lambda tup: tup[1], reverse=True)[:n]
        hits_h = [i[0] for i in hits_h]
        hits_a = sorted(tuple(zip(hits[1].keys(), hits[1].values())),key=lambda tup: tup[1], reverse=True)[:n]
        hits_a = [i[0] for i in hits_a]
    else:
        hits_h = None
        hits_a = None
    
    measures['dc'] = dcc
    measures['cc'] = ccc
    measures['bc'] = bcc
    measures['pr'] = prc
    measures['ha'] = hits_a
    measures['hh'] = hits_h
    return measures
```

```python
top_n_centrality_measures(G)
```
